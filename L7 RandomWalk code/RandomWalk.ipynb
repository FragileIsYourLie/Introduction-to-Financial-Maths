#### Simulating a Random Path

```python
def randomwalk(m):
    xx=np.random.choice(2,m)
    x=xx*2-1
    Y=[sum(x[:i]) for i in range(1,N+2)]
    Y.insert(0,0)
    return Y    
```

#### Ploting

```python
def plotrw(Y):
    m = len(Y)
    ts = np.arange(m)
    #fig,ax = plt.subplots(figsize = (7,5))
    # Use plt.step to plot piecewise constant functions
    plt.plot(ts,Y,color="darkred",lw=1)
    #plt.step(ts,Y,color="darkred",lw=1)
    plt.hlines(0,0,m+1,color="white",lw=1,alpha=.5)
    #plt.ylim([*K,1.1*K]) 
    plt.xlabel(r"time",fontsize = 13)
    #plt.ylabel(r"Random Walk",rotation =0,fontsize = 13,labelpad = 20) 
    plt.title("A Random Trajectory")
    plt.draw()
```

#### Example

```python
N=4000
Y= randomwalk(N)
plotrw(Y)
```

```python
N=1000
Y= randomwalk(N)
plotrw(Y)
```

```python
def simRWTerminal(m,K):
    #simulates K many values of the random walk at time m
    sim=[]
    for l in range(K):
        u=np.random.choice(2,m)
        Ym=2*sum(u)-m
        sim.append(Ym)
    return sim
```

```python
m=100
K=100000

sim = simRWTerminal(m,K)
xx=np.arange(m+1)
x=2*xx-m

plt.hist(sim,x)
#plt.title("m=",m)
plt.show()
```

```python
def reflect(Y,B):
    d = len(Y)
    H=0
    for k in range(1,d):
        if Y[k] <= B :
            H=H+1
        if Y[k] > B:
            break
            
    Yr = [0]*(d)
    for k in range(1,d):
        if k <= H:
            Yr[k]=Y[k]
        if k > H :
            Yr[k] = 2*B-Y[k]      
    return Yr  
```

```python
def plotboth(Y,B):
    m = len(Y)
    Yr = reflect(Y,B)
    ts = np.arange(m)
    #fig,ax = plt.subplots(figsize = (7,5))
    # Use plt.step to plot piecewise constant functions
    plt.plot(ts,Y,color="darkred",lw=1, label="original")
    plt.plot(ts,Yr,color="green",lw=1, label="reflected")
    #plt.step(ts,Y,color="darkred",lw=1)
    plt.hlines(0,0,m+1,color="white",lw=1,alpha=.5)
    plt.hlines(B,0,m+1,color="white",lw=1,alpha=.5)
    #plt.ylim([*K,1.1*K]) 
    plt.xlabel(r"time",fontsize = 13)
    #plt.ylabel(r"Random Walk",rotation =0,fontsize = 13,labelpad = 20) 
    plt.title("A Random Trajectory")
    plt.legend(loc = "upper right")
    plt.draw()
```

```python
N=50
Y= randomwalk(N)
plotboth(Y,0)
```
